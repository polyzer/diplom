```python
def ABS_to_XYZ(in_dir, out_dir):
    filenames = glob.glob(os.path.join(in_dir, "*.abs"))
    for fname in filenames:
        file_name = os.path.basename(fname)
        data = np.loadtxt(fname=fname, skiprows=3)
        data_transposed = data.transpose()
        data_cleared = data_transposed[np.where(data[0] == 1)]
        can_save = np.delete(data_cleared, 0, 1)
        save_name = os.path.join(out_dir, file_name + ".xyz")
        np.savetxt(fname=save_name, X=can_save)
        print(save_name)
```
